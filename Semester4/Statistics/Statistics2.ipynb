# **Метод инверсии гипотез для построения доверительного интервала разности средних**

## 1. Исходные данные и постановка задачи

Рассматриваются две независимые нормально распределённые выборки с одинаковой дисперсией:
$$
X_1 \sim \mathcal{N}(\mu_1, \sigma^2), \quad X_2 \sim \mathcal{N}(\mu_2, \sigma^2),
$$
где:
*   $\mu_1, \mu_2$ — неизвестные математические ожидания;
*   $\sigma^2$ — неизвестная, но одинаковая дисперсия в обеих выборках.

Требуется построить доверительный интервал для разности средних:
$$
\tau = \mu_1 - \mu_2.
$$

## 2. Статистика для проверки гипотез

Используем следующую статистику:
$$
T = \sqrt{\frac{n_1 n_2 (n_1 + n_2 - 2)}{n_1 + n_2}} \cdot \frac{\overline{X}_1 - \overline{X}_2 - \tau}{\sqrt{n_1 S^2_*(X_1) + n_2 S^2_*(X_2)}},
$$
где:
*   $n_1, n_2$ — объёмы выборок;
*   $\overline{X}_1, \overline{X}_2$ — выборочные средние;
*   $S^2_*(X_i) = \dfrac{1}{n_i}\sum\limits_{j=1}^{n_i}(X_{ij} - \overline{X}_i)^2$ — смещённая выборочная дисперсия.

## 3. Связь с $t$-распределением

Статистика $T$ имеет распределение Стьюдента с $\nu = n_1 + n_2 - 2$ степенями свободы:
$$
T \sim t(\nu).
$$

Эту формулу можно выразить из t-критерия Стьюдента для двух выборок с одинаковой дисперсия. И вот она имеет $T \sim t(\nu)$.

## 4. Проверка гипотез и доверительный интервал

Рассматриваем гипотезу:
$$
H_0: \tau = \tau_0 \quad \text{против} \quad H_1: \tau \neq \tau_0.
$$

Доверительный интервал уровня $1-\alpha$ включает все значения $\tau_0$, при которых гипотеза $H_0$ не отвергается:
$$
P\left(-t_{\text{крит}} \leq T \leq t_{\text{крит}}\right) = 1 - \alpha.
$$

Как мы докатились до жизни такой:
1.  Мы берем нашу T, которая показывает на самом деле грубо говоря сильно ли отклонилась $\tau$ из выборки от действительной $\tau$ ...
2.  ... и сравниваем ее с критическими значениями $-t_{\text{крит}}$. Шанс того что наша гипотеза не выполниться равна как раз 0.05

## 5. Критическое значение $t_{\text{крит}}$

Критическое значение определяется как:
$$
t_{\text{крит}} = t_{1 - \alpha/2, \nu},
$$
где:
*   $\alpha$ — уровень значимости;
*   $\nu = n_1 + n_2 - 2$ — число степеней свободы.

### Пояснение:
1.  При двусторонней проверке $\alpha$ делится пополам;
2.  Значение $t_{1 - \alpha/2, \nu}$ соответствует правому квантилю;
3.  Симметрия $t$-распределения даёт левый квантиль $-t_{\text{крит}}$.

## 6. Неравенство для границ интервала

Имеем:
$$
-t_{\text{крит}} \leq \sqrt{\frac{n_1 n_2 \nu}{n_1 + n_2}} \cdot \frac{\overline{X}_1 - \overline{X}_2 - \tau}{\sqrt{n_1 S^2_*(X_1) + n_2 S^2_*(X_2)}} \leq t_{\text{крит}}.
$$

Обозначим:
$$
C = \sqrt{\frac{n_1 n_2 \nu}{n_1 + n_2}}, \quad D = \sqrt{n_1 S^2_*(X_1) + n_2 S^2_*(X_2)}.
$$

Тогда неравенство:
$$
-t_{\text{крит}} \leq \frac{C}{D}(\overline{X}_1 - \overline{X}_2 - \tau) \leq t_{\text{крит}}.
$$

## 7. Вывод границ интервала

### Левая граница:
$$
\frac{C}{D}(\overline{X}_1 - \overline{X}_2 - \tau) \geq -t_{\text{крит}} \Rightarrow \tau \leq \overline{X}_1 - \overline{X}_2 + \frac{D}{C} t_{\text{крит}}.
$$

### Правая граница:
$$
\frac{C}{D}(\overline{X}_1 - \overline{X}_2 - \tau) \leq t_{\text{крит}} \Rightarrow \tau \geq \overline{X}_1 - \overline{X}_2 - \frac{D}{C} t_{\text{крит}}.
$$

### Итоговый доверительный интервал:
$$
\left[\overline{X}_1 - \overline{X}_2 - \frac{D}{C}t_{\text{крит}}, \ \overline{X}_1 - \overline{X}_2 + \frac{D}{C}t_{\text{крит}}\right].
$$

## 8. Роль уровня значимости $\alpha$

*   При $\alpha = 0.05$ доверительный уровень — 95%;
*   При уменьшении $\alpha$ критическое значение $t_{\text{крит}}$ увеличивается:
$$
t_{\text{крит}} \uparrow \quad \text{при} \quad \alpha \downarrow;
$$
*   Интервал расширяется при более строгих критериях;
*   Геометрически: площадь между $-t_{\text{крит}}$ и $t_{\text{крит}}$ под кривой $t$-распределения — $1 - \alpha$.

## 9. Как изменяется результат с ростом n
На самом деле у нас просто уменьшается наш интервал, причем довольно значительно, если глядеть на графики, так ваще экспонециально, не смотря на это у нас всё равно с вероятностью 0.95 $\tau$ с шапкой ушанкой все равно назодиться в доверительном интервале. Объяснение простое, на бытовом уровне, у нас $\mu$ стремятся к тем что указаны в теории, поэтому более точно и получается. Что еще тут писать я не знаю

```python
import numpy as np
from scipy.stats import t
from scipy.optimize import root_scalar
import matplotlib.pyplot as plt
import pandas as pd

def confidence_interval_by_hypothesis_inversion(X1, X2, alpha=0.05):
    """Построение доверительного интервала через инверсию проверки гипотез"""
    n1, n2 = len(X1), len(X2)
    X1_bar, X2_bar = np.mean(X1), np.mean(X2)
    S1_sq = np.var(X1, ddof=0)
    S2_sq = np.var(X2, ddof=0)


    C = np.sqrt((n1 * n2 * (n1 + n2 - 2)) / (n1 + n2))
    denominator = np.sqrt(n1 * S1_sq + n2 * S2_sq)
    t_crit = t.ppf(1 - alpha / 2, n1 + n2 - 2)


    def compute_T(tau):
        return C * (X1_bar - X2_bar - tau) / denominator


    def find_root(f, initial_guess, search_range=5):
        try:
            result = root_scalar(f, bracket=[initial_guess - search_range,
                                             initial_guess + search_range])
            return result.root
        except ValueError:
            return find_root(f, initial_guess, search_range * 2)

    point_estimate = X1_bar - X2_bar
    lower = find_root(lambda tau: compute_T(tau) + t_crit, point_estimate)
    upper = find_root(lambda tau: compute_T(tau) - t_crit, point_estimate)

    return upper, lower, point_estimate



n_sim = 1000
sample_sizes = np.linspace(25, 10000, 10, dtype=int)
true_tau = 1


results = {
    'n': [],
    'coverage': [],
    'mean_abs_error': [],
    'mean_width': []
}

for n in sample_sizes:
    coverage = 0
    abs_errors = []
    widths = []

    for _ in range(n_sim):
        X1 = np.random.normal(2, 1, n)
        X2 = np.random.normal(1, 1, n)
        lower, upper, tau_hat = confidence_interval_by_hypothesis_inversion(X1, X2)

        abs_errors.append(abs(tau_hat - true_tau))
        widths.append(upper - lower)

        if lower <= true_tau <= upper:
            coverage += 1

    results['n'].append(n)
    results['coverage'].append(coverage / n_sim)
    results['mean_abs_error'].append(np.mean(abs_errors))
    results['mean_width'].append(np.mean(widths))

# Визуализация
plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
plt.plot(results['n'], results['coverage'], 'o-')
plt.axhline(0.95, color='r', linestyle='--')
plt.xlabel('Размер выборки (n)')
plt.ylabel('Доля покрытия τ = 1')
plt.title('Покрытие истинного значения τ')
plt.grid(True)

plt.subplot(2, 2, 2)
plt.plot(results['n'], results['mean_abs_error'], 'o-')
plt.xlabel('Размер выборки (n)')
plt.ylabel('Средняя абсолютная ошибка')
plt.title('Точность оценки τ')
plt.grid(True)

plt.subplot(2, 2, 3)
plt.plot(results['n'], results['mean_width'], 'o-')
plt.xlabel('Размер выборки (n)')
plt.ylabel('Средняя ширина ДИ')
plt.title('Ширина доверительного интервала')
plt.grid(True)

plt.tight_layout()
plt.show()

# Таблица результатов
df = pd.DataFrame(results)
print(df[['n', 'coverage', 'mean_abs_error', 'mean_width']].round(4))
```

## Задание 2, Вариант 2: Асимптотический доверительный интервал для медианы распределения Лапласа

### Постановка задачи

Имеем выборку $X = (X_1, ..., X_n)$ из распределения Лапласа с параметром сдвига $\mu$ и единичным параметром масштаба ($b=1$).
Плотность распределения: $f(x; \mu) = \frac{1}{2b} e^{-|x-\mu|/b} = \frac{1}{2} e^{-|x-\mu|}$.
Параметр $\mu$ неизвестен.
Цель: построить **асимптотический** ДИ уровня $1 - \alpha$ для **медианы** этого распределения.

### Медиана и ее оценка

Для симметричного распределения Лапласа медиана совпадает с параметром сдвига $\mu$. То есть, мы ищем ДИ для $\mu$.
Естественной оценкой для медианы генеральной совокупности является **выборочная медиана**, обозначим ее $\hat{\mu} = \text{median}(X_1, ..., X_n)$.

### Асимптотическая теория

Для построения асимптотического ДИ нам нужна асимптотическая нормальность оценки.
**Теорема об асимптотическом распределении выборочных квантилей:** Для выборки из распределения с непрерывной и положительной в точке медианы $\mu$ плотностью $f(x)$, выборочная медиана $\hat{\mu}$ асимптотически нормальна при $n \to \infty$:
$$ \sqrt{n}(\hat{\mu} - \mu) \xrightarrow{d} N\left(0, \frac{1}{[2f(\mu)]^2}\right) $$
Здесь $\xrightarrow{d}$ означает сходимость по распределению.

**Применение к Лапласу:** Нам нужно найти значение плотности $f(x; \mu)$ в точке медианы $x = \mu$.
$$ f(\mu; \mu) = \frac{1}{2} e^{-|\mu-\mu|} = \frac{1}{2} e^0 = \frac{1}{2} $$
**Асимптотическая дисперсия:** Подставляем $f(\mu) = 1/2$ в формулу для дисперсии:
$$ \frac{1}{[2f(\mu)]^2} = \frac{1}{[2 \times (1/2)]^2} = \frac{1}{1^2} = 1 $$
**Асимптотическое распределение:** Таким образом, для выборочной медианы из распределения Лапласа$(\mu, 1)$:
$$ \sqrt{n}(\hat{\mu} - \mu) \xrightarrow{d} N(0, 1) $$
Это означает, что при больших $n$ распределение $\hat{\mu}$ можно аппроксимировать нормальным распределением $N(\mu, 1/n)$.

### Построение асимптотического доверительного интервала

Используем асимптотическую нормальность: $\frac{\hat{\mu} - \mu}{\sqrt{1/n}} \approx N(0, 1)$.
Берем квантили стандартного нормального распределения $z_{\alpha/2}$.
Приближенное вероятностное утверждение: $P\left(-z_{\alpha/2} \le \frac{\hat{\mu} - \mu}{1/\sqrt{n}} \le z_{\alpha/2}\right) \approx 1 - \alpha$.
Разрешаем неравенство относительно $\mu$:
$$ -z_{\alpha/2} \frac{1}{\sqrt{n}} \le \hat{\mu} - \mu \le z_{\alpha/2} \frac{1}{\sqrt{n}} $$
$$ \hat{\mu} - \frac{z_{\alpha/2}}{\sqrt{n}} \le \mu \le \hat{\mu} + \frac{z_{\alpha/2}}{\sqrt{n}} $$
Итоговый асимптотический $(1-\alpha)$-доверительный интервал для медианы $\mu$:
$$ \left[ \hat{\mu} - \frac{z_{\alpha/2}}{\sqrt{n}}, \quad \hat{\mu} + \frac{z_{\alpha/2}}{\sqrt{n}} \right] $$

```python
import numpy as np
from scipy import stats
import plotly.graph_objects as go
from plotly.subplots import make_subplots

def generate_laplace(mu, b, size):
    return np.random.laplace(mu, b, size)

def compute_confidence_interval(sample, alpha=0.05):
    n = len(sample)
    sample_median = np.median(sample)

    z_critical = stats.norm.ppf(1 - alpha/2)

    margin_of_error = z_critical / np.sqrt(n)

    return (sample_median - margin_of_error, sample_median + margin_of_error)


def run_simulation(n_samples, sample_size, true_mu=2, true_b=1, alpha=0.05):
    coverage_count = 0
    interval_lengths = []

    for _ in range(n_samples):
        sample = generate_laplace(true_mu, true_b, sample_size)

        ci_lower, ci_upper = compute_confidence_interval(sample, alpha)

        if ci_lower <= true_mu <= ci_upper:
            coverage_count += 1

        interval_lengths.append(ci_upper - ci_lower)

    coverage_rate = coverage_count / n_samples
    avg_interval_length = np.mean(interval_lengths)

    return coverage_rate, avg_interval_length

n_simulations = 1000
sample_sizes = np.linspace(25, 10000, 10, dtype=int)
true_mu = 2
true_b = 1
alpha = 0.05

results = []
for size in sample_sizes:
    coverage_rate, avg_length = run_simulation(
        n_simulations, size, true_mu, true_b, alpha
    )
    results.append({
        'size': size,
        'coverage': coverage_rate,
        'avg_length': avg_length
    })

print("\nРезультаты симуляции:")
print(f"Номинальный уровень доверия: {(1-alpha)*100}%")
print("\nРазмер выборки | Покрытие | Средняя длина интервала")
print("-" * 50)
for result in results:
    print(
        f"{result['size']:13d} | {result['coverage']*100:8.2f}% | {result['avg_length']:21.4f}"
    )

fig = make_subplots(
    rows=1,
    cols=2,
    subplot_titles=(
        'Уровень покрытия vs Размер выборки',
        'Длина интервала vs Размер выборки'
    )
)

fig.add_trace(
    go.Scatter(
        x=[r['size'] for r in results],
        y=[r['coverage'] for r in results],
        mode='lines+markers',
        marker=dict(color='blue'),
        name='Уровень покрытия'
    ),
    row=1,
    col=1
)

fig.add_trace(
    go.Scatter(
        x=sample_sizes,
        y=[1 - alpha] * len(sample_sizes),
        mode='lines',
        line=dict(color='red', dash='dash'),
        name='Номинальный уровень'
    ),
    row=1,
    col=1
)

fig.add_trace(
    go.Scatter(
        x=[r['size'] for r in results],
        y=[r['avg_length'] for r in results],
        mode='lines+markers',
        marker=dict(color='green'),
        name='Средняя длина интервала'
    ),
    row=1,
    col=2
)

fig.update_layout(
    title_text='Асимптотический доверительный интервал для медианы',
    xaxis1_title='Размер выборки',
    yaxis1_title='Уровень покрытия',
    xaxis2_title='Размер выборки',
    yaxis2_title='Средняя длина интервала',
    height=500,
    width=1200,
    showlegend=True
)

fig.update_xaxes(showgrid=True, row=1, col=1)
fig.update_yaxes(showgrid=True, row=1, col=1)
fig.update_xaxes(showgrid=True, row=1, col=2)
fig.update_yaxes(showgrid=True, row=1, col=2)

fig.show()
```
